```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Set up the figure and axis for multiple subplots
plt.figure(figsize=(14, 10))

# Bar plot for sepal length
plt.subplot(2, 2, 1)
sns.barplot(x="species", y="sepal_length", hue="species", data=iris, palette="Set2", legend=False)
plt.title("Average Sepal Length")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")

# Bar plot for sepal width
plt.subplot(2, 2, 2)
sns.barplot(x="species", y="sepal_width", hue="species", data=iris, palette="Set2", legend=False)
plt.title("Average Sepal Width")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")

# Bar plot for petal length
plt.subplot(2, 2, 3)
sns.barplot(x="species", y="petal_length", hue="species", data=iris, palette="Set2", legend=False)
plt.title("Average Petal Length")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")

# Bar plot for petal width
plt.subplot(2, 2, 4)
sns.barplot(x="species", y="petal_width", hue="species", data=iris, palette="Set2", legend=False)
plt.title("Average Petal Width")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")

# Adjust the layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Set up a grid for multiple plots
g = sns.FacetGrid(iris.melt(id_vars="species"), col="variable", col_wrap=2, sharex=False, sharey=False)
g.map(sns.barplot, "species", "value", "species", order=iris["species"].unique(), palette="Set2", ci=None)

# Set titles and labels
for ax, title in zip(g.axes.flatten(), ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]):
    ax.set_title(f"Average {title}")
    ax.set_xlabel("Species")
    ax.set_ylabel(f"{title} (cm)")

# Adjust the layout to prevent overlap
g.fig.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Set up the figure and axis for multiple subplots
plt.figure(figsize=(14, 10))

# Features and titles for the subplots
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
titles = ['Average Sepal Length', 'Average Sepal Width', 'Average Petal Length', 'Average Petal Width']

for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    sns.barplot(x="species", y=feature, hue="species", data=iris, palette="Set2")
    plt.title(titles[i])
    plt.xlabel("Species")
    plt.ylabel(f'{feature.replace("_", " ").title()} (cm)')
    plt.grid(axis='y')

# Adjust the layout to prevent overlap
plt.tight_layout()
plt.show()
```
